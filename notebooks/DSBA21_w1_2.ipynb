## Clustering Nomad Cities

```python
# packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data
data = pd.read_csv('https://sds-aau.github.io/SDS-master/M1/data/cities.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
# numerical variables
data.iloc[:,4:]
```

```python
# load up scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# scale the data (standardize)
data_scaled = scaler.fit_transform(data.iloc[:,4:])
```

```python
import numpy as np
```

```python
# load up PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
# perform PCS
data_pca = pca.fit_transform(data_scaled)
```

```python
data_pca.shape
```

```python
pca.components_.shape
```

```python
# plot the pca reduced data
sns.scatterplot(data_pca[:,0], data_pca[:,1], hue=data.region)
```

```python
sns.heatmap(pca.components_, cmap=sns.color_palette("rocket"))
```

### Introducing UMAP and Clustering with K-Means

```python
!pip install -q umap-learn
```

```python
import umap
reducer = umap.UMAP()
```

```python
umap_data = reducer.fit_transform(data_scaled)
umap_data.shape
```

```python
# Let's plot the matrix

sns.scatterplot(umap_data[:,0], umap_data[:,1], hue = data.region )
```

```python
from sklearn.cluster import KMeans
```

```python
clusterer = KMeans(n_clusters=3)
clusterer.fit(data_scaled)
```

```python
sns.scatterplot(umap_data[:,0], umap_data[:,1], hue = clusterer.labels_)
```

### Interactive Visualizing with Altair

```python

vis_data = pd.DataFrame(umap_data)
vis_data['cluster'] = clusterer.labels_
vis_data['place'] = data['place']
vis_data['country'] = data['alpha-2']
vis_data.columns = ['x', 'y', 'cluster', 'place', 'country']
```

```python
import altair as alt


alt.Chart(vis_data).mark_circle(size=60).encode(
    x='x',
    y='y',
    color='cluster',
    tooltip=['place', 'country']
).interactive()
```
